```python
import json

map_osm = folium.Map(location=[37.566345, 126.977893])

rfile = open('skorea_provinces_geo_simple.json', 'r', encoding='utf-8').read()



jsonData = json.loads(rfile)

folium.GeoJson(jsonData, name='json_data').add_to(map_osm)

map_osm.save('map7.html')
```

```python
import pandas as pd
```

```python
subway = pd.read_csv('./duranduranreturns_data/data/Subways.csv', encoding='utf-8')
```

```python
subway.head()
```

```python
map_1 = folium.Map(location=[45.372, -121.6972], tiles='Stamen Terrain', zoom_start=12)
folium.Marker([45.3288, -121.6625], popup='Mt. Hood Meadows').add_to(map_1)
folium.Marker([45.3311, -121.7113], popup='Timberline Lodge').add_to(map_1)
map_1
```

```python
map_osm = folium.Map(location=[37.566345, 126.977893])

rfile = open('skorea_provinces_geo_simple.json', 'r', encoding='utf-8').read()

for i in range(len(subway)):
    folium.Marker([subway['latitude'][i],subway['longitude'][i]]).add_to(map_osm)


jsonData = json.loads(rfile)

folium.GeoJson(jsonData, name='json_data').add_to(map_osm)

map_osm
```

```python
df_train['tallest_building_in_sites']=df_train['tallest_building_in_sites'].fillna(100)
df_train['lowest_building_in_sites']=df_train['lowest_building_in_sites'].fillna(100)
```

```python
df_tall_na=df_train[df_train['tallest_building_in_sites'] == 100]
df_tall_no_na=df_train[df_train['tallest_building_in_sites'] != 100]
```

```python
max(df_train['lowest_building_in_sites']),min(df_train['lowest_building_in_sites']),
```

```python
df_tall_no_na1=df_tall_no_na[['tallest_building_in_sites','year_of_completion','floor']]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X, y = df_tall_no_na1.iloc[:, 1:], df_tall_no_na1.iloc[:, 0]
lr.fit(X, y)
```

```python
print(lr.coef_,lr.intercept_)
```

```python

df_tall_na['tallest_building_in_sites']=round(((0.41203884   *df_tall_na['year_of_completion'])+(0.53147186 *df_tall_na['floor'])-808.1296944273432))
df_tall_na
```

```python
df_train=pd.merge(df_train,df_tall_na,how='left',on=['transaction_year', 'transaction_month', 'transaction_date1', 'key',
       'apartment_id', 'city', 'year_of_completion', 'exclusive_use_area',
       'floor', 'latitude', 'longitude', 'address_by_law',
       'total_parking_capacity_in_site', 'total_household_count_in_sites',
       'apartment_building_count_in_sites', 'tallest_building_in_sites',
       'lowest_building_in_sites', 'heat_type', 'heat_fuel', 'room_id',
       'supply_area', 'total_household_count_of_area_type', 'room_count',
       'bathroom_count', 'front_door_structure', 'transaction_real_price',
        'household_loan', 'mean_loan', 'mortgage loan'])
```

```python
df_low_na=df_train[df_train['lowest_building_in_sites'] == 100]
df_low_no_na=df_train[df_train['lowest_building_in_sites'] != 100]
```

```python
df_low_no_na1=df_low_no_na[['lowest_building_in_sites','floor','tallest_building_in_sites']]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X, y = df_low_no_na1.iloc[:, 1:], df_low_no_na1.iloc[:, 0]
lr.fit(X, y)
```

```python
print(lr.coef_,lr.intercept_)
```

```python

df_low_na['lowest_building_in_sites']=round(((0.06618384    *df_low_na['floor'])+(0.52759477 *df_low_na['tallest_building_in_sites'])-2.305682717585819))
df_low_na
```

```python
df_train=pd.merge(df_train,df_low_na,how='left',on=['transaction_year', 'transaction_month', 'transaction_date1', 'key',
       'apartment_id', 'city', 'year_of_completion', 'exclusive_use_area',
       'floor', 'latitude', 'longitude', 'address_by_law',
       'total_parking_capacity_in_site', 'total_household_count_in_sites',
       'apartment_building_count_in_sites', 'tallest_building_in_sites',
       'lowest_building_in_sites', 'heat_type', 'heat_fuel', 'room_id',
       'supply_area', 'total_household_count_of_area_type', 'room_count',
       'bathroom_count', 'front_door_structure', 'transaction_real_price',
        'household_loan', 'mean_loan', 'mortgage loan'])
```

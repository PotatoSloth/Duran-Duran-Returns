```python
df_total1=df_total.drop(['latitude','longitude'],axis=1)

df_train1=df_total1[df_total1['transaction_real_price']!=0]
df_test1=df_total1[df_total1['transaction_real_price']==0]
```

```python
df_total1.columns
```

```python
df_train1[['transaction_real_price']].head()
```

```python
train_data=df_train1.loc[:,'apartment_id':'subway']
train_target = df_train1[['transaction_real_price']]
test_data=df_test1.loc[:,'apartment_id':'subway']
test_target = df_test1[['transaction_real_price']]
```

```python
train_data= train_data.as_matrix()
train_target = train_target.as_matrix()
test_data = test_data.as_matrix()
test_target = test_target.as_matrix()
```

```python
# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옵니다.

from keras.models import Sequential
from keras.layers import Dense

import keras.models
import keras.layers

# 필요한 라이브러리를 불러옵니다.

import numpy
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분입니다.

seed = 55
numpy.random.seed(seed)
tf.set_random_seed(seed)

# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장합니다.

X = train_data
Y = train_target

# 딥러닝 구조를 결정합니다(모델을 설정하고 실행하는 부분입니다).

model = Sequential()
model.add(keras.layers.Dense(128, input_shape=(24,), activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(1))

Y_=model.predict(X)


# 딥러닝을 실행합니다.

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=10, batch_size=10000) # 에폭디폴트 : 30

predict_price = model.predict(test_data, batch_size=100, verbose = 1)

# 결과를 출력합니다.

#print('\n Accuracy: %.4f' % (model.evaluate(X, Y)[1]))
print(predict_price)
```
